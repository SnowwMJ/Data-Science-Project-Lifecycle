```python
# Import required columns for data cleaning
import pandas as pd
import re

# Load dataset
df = pd.DataFrame(data)

# Function to round to the nearest valid UK size
def round_to_half(size):
    return round(size * 2) / 2  # Rounds to the nearest whole or half size

# Optimised function to clean and standardise shoe sizes
def standardise_shoe_size(size):
    if pd.isna(size):
        return None

    # Convert to string, remove unwanted characters, and ensure consistent separators
    size = re.sub(r"[^0-9.,/ -]", "", str(size)).replace("/", ",").replace("-", ",").replace("&", ",").replace(" ", "")

    # Ensure proper splitting of multiple sizes
    sizes = re.split(r",+", size)  # Splits by one or more commas

    # Process valid numbers and convert EU sizes where necessary
    valid_sizes = []
    for s in sizes:
        s = s.strip().rstrip(".")  # Clean up each value
        if s.replace(".", "").isdigit():  # Ensure it's a valid number
            try:
                num = float(s)
                if 20 < num < 50:  # Likely an EU size, convert to UK
                    uk_size = round_to_half(num - 33)
                    if 1.0 <= uk_size <= 13.5:
                        valid_sizes.append(str(uk_size))
                elif 1.0 <= num <= 13.5:  # Valid UK size
                    valid_sizes.append(str(round_to_half(num)))
            except ValueError:
                continue  # Skip invalid entries

    return valid_sizes if valid_sizes else None

# Apply function and get unique valid sizes
df["Standardized Shoe Sizes"] = df['Shoe Size (UK) \nPlease enter a UK shoe size. '].apply(standardise_shoe_size)
unique_sizes = sorted(set(size for sublist in df["Standardized Shoe Sizes"].dropna() for size in sublist))

# Display results
print("Unique Standardized Shoe Sizes:", unique_sizes)
print(df[["Shoe Size (UK) \nPlease enter a UK shoe size. ", "Standardized Shoe Sizes"]])
```

```python
# Calculate the mean shoe size (convert list of strings to floats)
mean_shoe_size = pd.to_numeric(df["Standardized Shoe Sizes"].dropna().explode()).mean()

# Round to the nearest whole or half size
mean_shoe_size = round(mean_shoe_size * 2) / 2

print("Mean Shoe Size (Rounded to Whole/Half):", mean_shoe_size)
```

```python
import pandas as pd

# Ensure Timestamp column is in datetime format (day/month/year format explicitly set)
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%d/%m/%Y", errors='coerce')

# Extract year and month for analysis
df["Year"] = df["Timestamp"].dt.year
df["Month"] = df["Timestamp"].dt.month

# Count applicants per year
yearly_counts = df["Year"].value_counts().sort_index()
print("Applicants Per Year:\n", yearly_counts)

# Count applicants per month in each year
monthly_distribution = df.groupby(["Year", "Month"]).size().unstack(fill_value=0)
print("\nMonthly Distribution Per Year:\n", monthly_distribution)
```

```python
import pandas as pd

# Ensure the Ethnicity column is treated as a string and fill NaN with "Unknown"
df["Ethnicity "] = df["Ethnicity "].astype(str).str.strip().replace("nan", "Unknown")

# Normalise text (e.g., lowercase for consistent counting)
df["Ethnicity "] = df["Ethnicity "].str.lower().str.replace("-", " ").str.replace("/", " ").str.strip()

# Count frequency of each unique value
ethnicity_counts = df["Ethnicity "].value_counts()

# Display all unique inputs with frequencies
print("All Unique Ethnicities with Counts:\n", ethnicity_counts)

# Get the 3 most common demographics
top_3_ethnicities = ethnicity_counts.head(3)
print("\nTop 3 Most Common Ethnicities:\n", top_3_ethnicities)
```
