```python
df_cleaned = pd.read_csv('/content/Shareware_Data.csv', skiprows=1, header=None) # ingesting the data into here

# renaming columns using the first row as headers
df_cleaned.columns = df_cleaned.iloc[0]
df_cleaned = df_cleaned[1:].reset_index(drop=True)

# converting numerical columns as categoricals cant be visulised
numeric_columns = [
    "Total Clothes Distributed",
    "Adult Clothes Distributed",
    "Kids Clothes Distributed",
    "Counties Supported",
    "Items Chosen Post-Processing",
    "Items Processed (in tons)"
]

# converts all the values in the numeric columns to numeric data types sepcified abive
df_cleaned[numeric_columns] = df_cleaned[numeric_columns].apply(pd.to_numeric)

# convertig months column to datetime format so it odesnt give probelms in powerbi
df_cleaned["Month"] = pd.to_datetime(df_cleaned["Month"], format="%m/%y")

df_cleaned.head()
```

```python
df_cleaned.to_csv(r'/content/New_Shareware_Data.csv', index=False) # saving the cleaned data
```
